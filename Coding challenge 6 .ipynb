**Project Title: Shopping Cart System**

**Scenario:**

You are creating a simple billing system where customers can enter the prices of items they want to purchase. The program should keep accepting prices until the user decides to stop, then apply appropriate discounts before generating the final bill.
**Tasks:**

Continuously accept item prices using a while loop.
Stop taking input when the user enters 0 (sentinel value).
Keep track of the total number of items and the subtotal amount.
Ensure no negative prices are allowed (ask again without counting as an item).

Apply discounts based on the subtotal:
20% for totals ≥ 1000
10% for totals ≥ 500
5% for totals ≥ 200
0% for totals < 200

Display a clear bill summary including subtotal, discount amount, and final total. in simple code

```python
# Shopping Cart System

subtotal = 0
item_count = 0

while True:
    price = float(input("Enter item price (0 to stop): "))

    # Stop if user enters 0
    if price == 0:
        break

    # Reject negative price
    if price < 0:
        print("Negative price not allowed. Please enter again.")
        continue

    # Add to subtotal and count items
    subtotal += price
    item_count += 1

# Apply discounts
if subtotal >= 1000:
    discount_rate = 0.20
elif subtotal >= 500:
    discount_rate = 0.10
elif subtotal >= 200:
    discount_rate = 0.05
else:
    discount_rate = 0.0

discount_amount = subtotal * discount_rate
final_total = subtotal - discount_amount

# Bill Summary
print("\n----- BILL SUMMARY -----")
print("Total Items      :", item_count)
print("Subtotal Amount  : ₹", subtotal)
print("Discount Applied :", discount_rate * 100, "%")
print("Discount Amount  : ₹", discount_amount)
print("Final Total      : ₹", final_total)
print("-------------------------")
```
