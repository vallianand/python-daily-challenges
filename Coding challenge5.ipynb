 **Coding challenge 5**
 Dictionary, Set, Frozenset
Project Title: Student Grades & Hobbies Tracker
Scenario:
 A student activity coordinator is collecting information about student grades and hobbies. They want to ensure grades are updated properly and hobby lists are stored efficiently.
Tasks:
Create a dictionary with student names as keys and their grades as values.


Allow the user to add or update grades for at least 3 students.


Create a set of hobbies and allow the user to input at least 3 hobbies.


Convert the set of hobbies into a frozenset and display it.


```python
# Project Title: Student Grades & Hobbies Tracker

# Step 1: Create a dictionary with student names and grades (user input)
student_grades = {}

name1 = input("Enter first student name: ")
grade1 = input(f"Enter {name1}'s grade: ")

name2 = input("Enter second student name: ")
grade2 = input(f"Enter {name2}'s grade: ")

name3 = input("Enter third student name: ")
grade3 = input(f"Enter {name3}'s grade: ")

# Add all to dictionary
student_grades[name1] = grade1
student_grades[name2] = grade2
student_grades[name3] = grade3

# Display the dictionary
print("\nStudent Grades:")
print(student_grades)

# Step 2: Create a set of hobbies (user input)
hobby1 = input("\nEnter first hobby: ")
hobby2 = input("Enter second hobby: ")
hobby3 = input("Enter third hobby: ")

hobbies = {hobby1, hobby2, hobby3}

# Display the set
print("\nHobbies Set:", hobbies)

# Step 3: Convert the set into a frozenset
frozen_hobbies = frozenset(hobbies)

# Display the frozenset
print("Frozenset of Hobbies:", frozen_hobbies)
```
