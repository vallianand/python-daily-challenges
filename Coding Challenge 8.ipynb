**Built-in Functions and User-Defined Functions**

**Project Title: Simple Calculator App**

**Scenario:**
 A company wants to provide users with a simple calculator to perform basic mathematical operations.

**Tasks:**

Create functions for addition, subtraction, multiplication, and division.
Ask the user to input two numbers and the operation they want to perform.
Call the appropriate function and display the result.
Use built-in functions like abs(), round(), and pow() within your solution.
Handle division by zero gracefully and inform the user.

```python
# Simple Calculator App
# Topic: Built-in Functions & User-Defined Functions

# User-defined functions
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Division by zero is not allowed!"
    return a / b

print("ðŸ“Œ Simple Calculator App")
print("------------------------")

# Taking user inputs
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Choose operation (+, -, *, /): ")

# Calling appropriate function
if operation == "+":
    result = add(num1, num2)

elif operation == "-":
    result = subtract(num1, num2)

elif operation == "*":
    result = multiply(num1, num2)

elif operation == "/":
    result = divide(num1, num2)

else:
    result = "Invalid operation selected!"

print("\nðŸ“˜ Result:", result)

# Using built-in functions
print("\nðŸ”§ Built-in Function Outputs:")
print("Absolute value of result:", abs(result) if isinstance(result, (int, float)) else result)
print("Rounded value:", round(result, 2) if isinstance(result, (int, float)) else result)
print("Square using pow():", pow(result, 2) if isinstance(result, (int, float)) else result)
```
